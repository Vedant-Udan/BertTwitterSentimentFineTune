```python
import pandas as pd
```

```python
df = pd.read_csv('Sentiment Analysis Dataset.csv', error_bad_lines=False)
df.head()
```

```python
df.info()
```

```python
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
import string
```

```python
print(string.punctuation)
```

```python
stopword = set(stopwords.words('english'))
print(stopword)
```

```python
# Few more stopwords that are used in twitter.
additional_stopwords = {'amp', 'rt', 'via', '...', '...', 'u', '2', 'im', 'dont', "i'm", 'youre', "you're", 'ur', 'r', 'ive', "i've", 'id', "i'd", 'im', "i'm"}
final_stopwords = stopword.union(additional_stopwords)
```

```python
import re
def preprocessing(raw_text):
  regex = re.compile('[%s0-9]' % re.escape(string.punctuation))
  raw_text = regex.sub(' ', raw_text)
  processed_text = " ".join([x.lower() for x in raw_text.split(" ") if x.lower() not in final_stopwords and x != " "])
  return processed_text.strip()
```

```python
processed_df = df.copy()
processed_df = processed_df[:50000]
processed_df.head()
```

```python
processed_df['SentimentText'] = df['SentimentText'].map(preprocessing)
print(f'raw_df SentimentText: {df["SentimentText"][0]}')
print(f'processed_df SentimentText: {processed_df["SentimentText"][0]}')
```

```python
processed_df.tail()
```

```python
processed_df['SentimentText'][3]
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
max_len = 0

for sentence in processed_df['SentimentText']:
  tokens = tokenizer.encode(sentence, add_special_tokens=True)
  max_len = max(max_len, len(tokens))

print(max_len)
```

```python

input_ids = []
attentions_mask = []

for sentence in processed_df['SentimentText']:
  encoded_dict = tokenizer.encode_plus(
      sentence,
      add_special_tokens = True,
      max_length = max_len,
      pad_to_max_length = True,
      return_attention_mask = True,
      truncation=True,
      return_tensors = 'pt',
  )
  input_ids.append(encoded_dict['input_ids'])
  attentions_mask.append(encoded_dict['attention_mask'])
```

```python
input_ids[0]
```

```python
import torch
```

```python
input_ids = torch.cat(input_ids, dim = 0 )
attentions_mask = torch.cat(attentions_mask, dim=0)
labels = torch.tensor(processed_df.Sentiment)
```

```python
from torch.utils.data import TensorDataset, random_split, DataLoader
```

```python
dataset = TensorDataset(input_ids, attentions_mask, labels)

split_len = 0.9
train_size = int(split_len * len(dataset))
val_size = len(dataset) - train_size
train_data, val_data = random_split(dataset, [train_size, val_size])

train_loader = DataLoader(train_data, batch_size = 32)
val_loader = DataLoader(val_data, batch_size = 32)
```

```python
from transformers import BertForSequenceClassification, AdamW, BertConfig

model = BertForSequenceClassification.from_pretrained( 'bert-base-uncased', num_labels = 2, output_attentions = False, output_hidden_states = False )
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
model.to(device)
```

```python
num_epochs = 2
learning_rate = 3e-5
```

```python
optimizer = AdamW( model.parameters(), lr = learning_rate)
```

```python
for i in range(num_epochs):
  model.train()
  train_loss = 0
  for step, batch in enumerate(train_loader):
    input_id = batch[0].to(device)
    attention_mask = batch[1].to(device)
    label = batch[2].to(device)
    output = model(input_id,token_type_ids=None, attention_mask= attention_mask, labels = label)
    loss, logit = output[:2]
    train_loss += loss.item()
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

  print(f'training loss for epoch { i+1 } is {train_loss / len(train_loader)}')

  model.eval()
  val_loss = 0
  for step, batch in enumerate(val_loader):
    input_id = batch[0].to(device)
    attention_mask = batch[1].to(device)
    label = batch[2].to(device)
    with torch.no_grad():
      output = model(input_id,token_type_ids=None, attention_mask= attention_mask, labels = label)
      loss, logit = output[:2]
      val_loss += loss.item()
  print(f'validation loss for epoch { i+1 } is {val_loss / len(val_loader)}')
```
